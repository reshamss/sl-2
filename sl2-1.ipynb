```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate data for plotting
x = np.linspace(-10, 10, 400)

# Set the figure size for plots
plt.figure(figsize=(12, 8))
```

```python
def linear_activation(x):
    return x

x_values = range(-5, 5)
y_values = [linear_activation(x) for x in x_values]

plt.figure(figsize=(6, 4))  # Optional: Set figure size
plt.subplot(1, 1, 1)  # Only one plot, so a 1x1 grid
plt.plot(x_values, y_values)
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Linear Activation Function')
plt.grid(True)
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Plotting
plt.figure(figsize=(6, 4))  # Optional: adjust size
plt.plot(x, sigmoid(x), label='Sigmoid', color='blue')
plt.title('Sigmoid Activation Function')
plt.xlabel('Input')
plt.ylabel('Sigmoid Output')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)

# Plotting
plt.figure(figsize=(6, 4))  # Optional figure size
plt.plot(x, relu(x), label='ReLU', color='orange')
plt.title('ReLU Activation Function')
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

# Plotting
plt.figure(figsize=(6, 4))
plt.plot(x, tanh(x), label='Tanh', color='green')
plt.title('Tanh Activation Function')
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# Plotting
plt.figure(figsize=(6, 4))
plt.plot(x, leaky_relu(x), label='Leaky ReLU', color='red')
plt.title('Leaky ReLU Activation Function')
plt.xlabel("x")
plt.ylabel("Leaky ReLU(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
def softmax(x):
    e_x = np.exp(x - np.max(x))  # subtract max for numerical stability
    return e_x / np.sum(e_x)

# Plotting
plt.figure(figsize=(6, 4))
plt.plot(x, softmax(x), label='Softmax', color='blue')
plt.title('Softmax Activation Function')
plt.xlabel("x")
plt.ylabel("softmax(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
def elu(x, alpha=0.5):
    return [alpha * (np.exp(i) - 1) if i < 0 else i for i in x]

# Plotting
plt.figure(figsize=(6, 4))
plt.plot(x, elu(x), label='ELU', color='yellow')
plt.title('ELU Activation Function')
plt.xlabel("x")
plt.ylabel("ELU(x)")
plt.grid(True)
plt.legend()
plt.show()
```
